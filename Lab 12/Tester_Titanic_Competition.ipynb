```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
test_ids = test['PassengerId']
```

```python
df.head(5)
```

```python
def clean(df):
    df = df.drop(["Ticket", "Cabin", "Name", "PassengerId"], axis=1)
    
    cols = ["SibSp", "Parch", "Fare", "Age"]
    for i in cols:
        df[i] = df[i].fillna(df[i].median(), inplace=True)
    
    df["Embarked"] = df["Embarked"].fillna("S")
    return df

df = clean(df)
test = clean(test)
```

```python
df.head(5)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

cols = ["Sex", "Embarked"]

# Fitting the label encoder

for col in cols:
    df[col] = le.fit_transform(df[col])
    test[col] = le.transform(test[col])
    print(le.classes_) 

df.head(5)    
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df["Survived"]
Y = df.drop(["Survived"], axis=1)

X_train, X_test, y_train, y_test = train_test_split(Y, X, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(random_state=0, max_iter=1000)
clf.fit(X_train, y_train)
```

```python
Predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, Predictions)
```

```python
Submission_pred = clf.predict(test)
```

```python
data = pd.DataFrame({"PassengerId": test_ids.values,
                    "Survived": Submission_pred})
```
